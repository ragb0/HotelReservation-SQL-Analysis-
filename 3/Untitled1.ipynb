```python
# Calculate total cost for each city
columns_to_sum = [f'x{i}' for i in range(1, 56)]
data['TotalCost'] = data[columns_to_sum].sum(axis=1)

# Filter out rows with low data quality if necessary
data = data[data['data_quality'] == 1]

# Display first few rows to verify
print(data.head())
```

```python
# Columns related to salary, housing, and transportation
salary_col = 'x54'  # Average Monthly Net Salary
housing_cols = ['x48', 'x49', 'x50', 'x51']  # Housing costs columns
transportation_cols = ['x28', 'x29', 'x30', 'x31']  # Transportation costs columns

# Calculate average housing and transportation costs
data['AverageHousingCost'] = data[housing_cols].mean(axis=1)
data['AverageTransportationCost'] = data[transportation_cols].mean(axis=1)

# Display first few rows to verify
print(data[['city', 'country', 'TotalCost', 'AverageHousingCost', 'AverageTransportationCost', salary_col]].head())
```

```python
import matplotlib.pyplot as plt

# Sort data by TotalCost
sorted_data = data.sort_values(by='TotalCost', ascending=False)

# Top 10 and Bottom 10 cities by TotalCost
top_10 = sorted_data.head(10)
bottom_10 = sorted_data.tail(10)

# Plotting
plt.figure(figsize=(14, 7))

# Top 10
plt.subplot(1, 2, 1)
plt.barh(top_10['city'], top_10['TotalCost'], color='green')
plt.title('Top 10 Cities with Highest Cost of Living')
plt.xlabel('Total Cost')
plt.ylabel('City')

# Bottom 10
plt.subplot(1, 2, 2)
plt.barh(bottom_10['city'], bottom_10['TotalCost'], color='red')
plt.title('Top 10 Cities with Lowest Cost of Living')
plt.xlabel('Total Cost')
plt.ylabel('City')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Sum of cost components
cost_components = data[columns_to_sum].sum().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(14, 7))
sns.barplot(x=cost_components.values, y=cost_components.index, palette='viridis')
plt.title('Major Cost Components Contributing to the Overall Cost of Living')
plt.xlabel('Sum of Costs')
plt.ylabel('Cost Components')
plt.show()
```

```python
# Scatter plot to show correlation
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
sns.scatterplot(data=data, x=salary_col, y='AverageHousingCost')
plt.title('Correlation between Average Salary and Housing Costs')
plt.xlabel('Average Monthly Net Salary (USD)')
plt.ylabel('Average Housing Cost (USD)')

plt.subplot(1, 2, 2)
sns.scatterplot(data=data, x=salary_col, y='AverageTransportationCost')
plt.title('Correlation between Average Salary and Transportation Expenses')
plt.xlabel('Average Monthly Net Salary (USD)')
plt.ylabel('Average Transportation Cost (USD)')

plt.tight_layout()
plt.show()
```

```python
# Pair plot to see relationships between key variables
sns.pairplot(data, vars=['TotalCost', salary_col, 'AverageHousingCost', 'AverageTransportationCost'])
plt.show()
```

```python

# Correlation heatmap
plt.figure(figsize=(12, 8))
corr = data[['TotalCost', salary_col, 'AverageHousingCost', 'AverageTransportationCost']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
